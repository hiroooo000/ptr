```python
import matplotlib.pyplot as plt

#plt.style.use('seaborn-whitegrid')
import numpy as np
#%matplotlib inline

x = np.linspace(0, 10, 10000) #わからない1

#fig = plt.figure() #わからない2
#ax = plt.axes() #わからない3

plt.plot(x, np.sin(x)) #同じfigureの中に追加
plt.plot(x, np.cos(x)) #同じfigureの中に追加
```

# figureが1つの場合

```python
x = np.linspace(0, 10, 20)
y1 = x
y2 = x ** 2 

# figure は 1 つ
plt.figure(figsize=(6, 2)) # figureの縦横の大きさ　数字はインチ

plt.subplot(1,2,1) # figure内の枠の大きさとどこに配置している。subplot(行の数,列の数,何番目に配置しているか)
plt.plot(x, y1)

plt.subplot(1,2,2)
plt.plot(x, y2)
```

# figureが2つの場合

```python
x = np.linspace(-3, 3, 20)
y1 = x
y2 = x ** 2

# figure 1 つ目
plt.figure(figsize=(3, 2))
plt.plot(x, y1)

# figure 2 つ目
plt.figure(figsize=(3, 2))
plt.plot(x, y2)
```

# subplotsとax[x,x]


```python
x = np.linspace(-3, 3, 20)
y1 = x
y2 = x ** 2
y3 = x ** 3
y4 = x ** 4

fig, ax = plt.subplots(2, 2, figsize=(6, 4))

# 左上
ax[0, 0].plot(x, y1)

# 右上
ax[0, 1].plot(x, y2)

# 左下
ax[1, 0].plot(x, y3)

# 右下
ax[1, 1].plot(x, y4)
```
